# Непрозрачность I

2. "Положение о Совете обучающихся по основным общеобразовательным программам Лицея Национального исследовательского университета «Высшая школа экономики»" (Источник: https://www.hse.ru/docs/228930727.html) (от 2018 года)
Следовательно, мы будем использовать пункт 1. для дальнейшего обоснования, ввиду того, что это наиболее современная версия положения.

# Не Социально Полезный Проект

Совет Лицеистов - это большой (глава получает 15 баллов) социально полезный проект (Источник: https:\\2359.ru)  

1. Это противоречит "Положению о Совете Лицеистов" от 2021 года (Источник: https://school.hse.ru/tusalyc/samoupravlenie/sovetl). Вот цитата из положения:

> Учащиеся Лицея НИУ ВШЭ принимают участие в деятельности Совета лицеистов на безвозмездной основе. (пункт 1.4 положения от 2021 года)

Из данной цитаты мы можем отчётливо понять, что Совет Лицеистов не может быть СПД (Социально Полезная Деятельность) проектом по его же определению.

2. По критериям Социально Полезных проектов, у медиа проекта, должно быть минимум 4 поста в 1 неделю, у НЕ медиа проекта должен быть 1 пост в 1 неделю. За исключением выборного этапа, такой статистики не наблюдается (Источник: tgstat.ru)  
![alt](res/СЛ_публикации.png)  
То есть, мы можем сделать вывод, что Совет Лицеистов не медиа проект. Давайте рассмотрим события, которые организовал Совет Лицеистов и оценим насколько они социально-полезные. Но, мы не берём в расчёт выборы, так как, в данном кейсе, выборы в Совет Лицеистов - это отбор в СПД проект. Если бы каждый проект считал отбор в него как отдельное событие, то все проекты в Лицее были бы большими (100+ человек). Исходя из этого, вот список событий:  
    1. Пробное ИВР    
    2. Пробное ПДО    
    3. Лекторий Сейф Спейс     
    4. Гумный фестиваль    
Нельзя отрицать, что их события, особенно, пробное ПДО и СПД - это благородные дела. Однако, нельзя сказать, что проект будет существовать постоянно засчёт придуманной крайне давно проблематики. Тем более, что ПДО не имеет блокирующей способности и любой девятиклассник может поговорить про направление с учениками этого направления и или с руководителем данного направления напрямую. Лекторий Сейф Спейс - это довольно интересная задумка, но она не имеет проблематики и целевой аудитории, ввиду малого количества участников. То же самое и со станцией на Гумном фестивале.

3. По официальному определению Совета Лицеистов, Совет лицеистов - орган ученического самоуправления Лицея НИУ ВШЭ, нацеленный на защиту законных интересов лицеистов. Также, Совет Лицеистов не один раз говорил, что они - посредники между Лицеистами и Администрацией. Однако, это тоже крайне спорно:
    1. Лицеисты x Администрация:   
    Возможно, на момент создания Совета Лицеистов договоры с администрацией Лицея были крайне бюрократически сложными. Однако, в нынешнее время, все события организуются напрямую с администрацией ввиду того, что это быстрее и Лицеисты могут чётко обозначить чего они хотят.
    2. СПД проекты x Администрация:
    В Ноябре этого года, случилась крайне неприятная ситуация. С момента создания Лицея, разнообразные звонки были очень хорошей деятельностью. Возник проект Lyceum Bells, отвечающий за эти звонки. Но, в Ноябре 2024 года, звонки на здании Солянки перестали работать должным образом. Лицеисты не раз обращались в Совет за помощью. Им нужно было всего лишь позвать коменданта по зданию для констатации поломки коробки для звонков. Не смотря на свои обещания, Совет Лицеистов не смог решить данную проблему до марта 2025 года! Причём, председатель Совета Лицеистов всегда говорил, что это не их зона ответственности. Из его слов следует то, что существование АХК (Административно Хозяйственного) отдела Совета Лицеистов необоснованно и не имеет проблематики.
